# If you open up netflix the first thing you think about is what to watch, sometimes you even spend like 10 min or more just finiding what to watch. I wanted to see if there is anything that can help someone make the decision faster or help them get to know what netflix consists of to make it easier to choose. 

DATASET: https://www.kaggle.com/shivamb/netflix-shows

This Dataset is of Netflix, consists of what TV Shows and Movies there are, which country they are from, their duration, releasing year and much more.

```python
import pandas as pd

df=pd.read_csv("netflix_titles.csv")
df.head()
```

# <mark> Pie Chart

```python
percent = df['type'].value_counts()
percent
```

```python
values = percent.index
plt.pie(percent.values, labels = values,autopct="%1.1f%%")
plt.show()
```

## It is evident that there are more Movies on Netflix than TV shows.



# <mark> Line graph

```python
df.dtypes
```

```python
# to change from integer to string
df["release_year"]=df["release_year"].astype(str)
```

```python
years=df['release_year'].value_counts().head(18).sort_index(axis=0)
years
```

### Separate TV SHOWS AND MOVIES

```python
TVdf=df[df['type']=="TV Show"]
TVdf
```

```python
moviedf=df[df['type']=="Movie"]
moviedf
```

```python
years.plot.line()
plt.xlabel("Year")
plt.ylabel("Count")
plt.title('All Types of Releases yearly')
```

## In general,the amount of releases decreased after the year 2018. 


```python
TVyrs=TVdf['release_year'].value_counts().head(17).sort_index(axis=0)
yrs=moviedf['release_year'].value_counts().head(16).sort_index(axis=0)
```

```python
plt.plot(TVyrs, label = "TV")
plt.plot(yrs, label = "Movies")
plt.xticks(rotation=90)
plt.xlabel("Year")
plt.ylabel("Count")
plt.title('Releases Yearly')
plt.legend()
plt.show()
```

## Though the movies releases decreased but tv series at that time were increasing until 2020

# <mark> Histogram 

```python
# moviedf["duration"]=moviedf["duration"].astype(int)
moviedf['duration']=moviedf['duration'].str.replace(' min','')
moviedf['duration']=moviedf['duration'].astype(str).astype(int)
md=moviedf['duration']
```

```python
md.value_counts(bins=18, sort=True)
```

```python
md.plot.hist(bins=18)
plt.xlabel("Duration (min)")
plt.title('Duration of movies')
```

## we see that most movies are between 88 mins to 106 mins   

# <mark> Bar Graph

```python
# TOP 5 CONTRY FOR MOVIE:
country_count1 = pd.crosstab(moviedf["country"],["type"]).sort_values(by = 'type',ascending=False )
country_count1.head(5).plot(kind="bar",color ="black")
plt.title("TOP 5 COUNTRIES WITH HIGHEST NUMBER OF MOVIES")
plt.legend("Movies")
plt.show()
#TOP 5 COUNTRY FOR TV SERIES:
country_count2 = pd.crosstab(TVdf["country"],["type"]).sort_values(by = 'type',ascending=False )
country_count2.head(5).plot(kind="bar",color ="brown")
plt.legend("TV Shows")
plt.title("TOP 5 COUNTRIES WITH HIGHEST NUMBER OF TV SHOWS")
plt.show()
#TOP 5 COUNTRY OVERALL CONTENT ON NETFLIX:
country_count3 = pd.crosstab(df["country"],["type"]).sort_values(by = 'type',ascending=False )
country_count3.head(5).plot(kind="bar",color ="orange")
plt.legend("Content")
plt.title("TOP 5 COUNTRIES CONTENT ON NETFLIX")
plt.show()

x1=country_count1.head(5)
x2=country_count2.head(5)
x3=country_count3.head(5)
```

# We see that the top country for both movies and tv shows is U.S and we also see that though Inidia is 2nd for movies, it is 5th for TV shows. Showing that there are way more indian movies on netflix than indian TV shows.


# <mark>Scatterplot

```python
mr=moviedf[moviedf['release_year']>'2012']
x=mr['duration']
y=mr['release_year'].sort_values()
plt.scatter(x,y)
plt.xlabel('Duration')
plt.ylabel('Years')
plt.title("Duration over the years")
```

# we see that mostly the duration range every year are similar.Only in 2015 there was a very far outlier,showing one that was too long of duaration.

```python
x=mr['duration']
y=mr['release_year'].sort_values()

fig, axs = plt.subplots(3,2,figsize=(30,25))

# 1st 
axs[0, 0].pie(percent.values, labels = values,autopct="%1.1f%%")
axs[0, 0].set_title('MOVIES VS TV SHOWS')
#2nd
axs[0, 1].plot(years)
axs[0, 1].set_title('Total Releases yearly')
axs[0, 1].set_xlabel(' years')
axs[0, 1].set_ylabel('count')
#3rd
axs[1, 0].plot(TVyrs)
axs[1, 0].plot(yrs)
axs[1, 0].set_title('Releases Yearly                      Orange-Movies and Blue- TV')
axs[1, 0].set_xlabel(' years')
axs[1, 0].set_ylabel('count')
#4th
axs[1, 1].hist(md,bins=18)
axs[1, 1].set_title('Duration of movies')
axs[1,1].set_xlabel('Duration (min)')
axs[1, 1].set_ylabel('frequency')
#5th
axs[2,0].bar(x1.index,x3['type'])
axs[2,0].set_title('Top 5 Countries content on Netflix')
axs[2,0].set_xlabel('Country')
axs[2,0].set_ylabel('Count of movies')
#6th
axs[2,1].scatter(x,y)
axs[2,1].set_title('Duration over the years')
axs[2,1].set_xlabel('Duration')
axs[2,1].set_ylabel('years')

```
